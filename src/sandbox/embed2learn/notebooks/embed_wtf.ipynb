```python
import pickle
import os
import copy
import numpy as np
import matplotlib

# Avoid type 3 fonts
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

# WTF happened?! - analyzing samples of embedded policies...


s1 = os.path.join(os.getcwd(), "../../../play_dqn.pkl")
s2 = os.path.join(os.getcwd(), "../../../play_embed.pkl")

with open(s1, "rb") as fin:
    samples1 = pickle.load(fin)
with open(s2, "rb") as fin:
    samples2 = pickle.load(fin)

s1, s2, len(samples1["obs"]), len(samples2["obs"])
```

```python
import matplotlib.pyplot as plt

def plt_diff(what):
    plt.plot(np.array(samples1[what]) - np.array(samples2[what]))
    plt.title(what)
    print("diff %s:" % what, np.linalg.norm(np.array(samples1[what]) - np.array(samples2[what])))

plt_diff("latents")
```
